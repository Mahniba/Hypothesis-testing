```python
#user entries are allowed to calculate the z values
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha):
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Calculate the critical value from the normal distribution
    critical_value = norm.ppf(1 - alpha / 2)
    
    return z_score, critical_value

def plot_z_test(z_score, critical_value, alpha):
    # Plot the normal distribution curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Normal Distribution')
    
    # Shade the critical regions
    plt.fill_between(x, 0, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label='Critical Regions')
    plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-Score = {z_score:.2f}')
    plt.axvline(-critical_value, color='red', linestyle='--', linewidth=2)
    plt.axvline(critical_value, color='red', linestyle='--', linewidth=2)
    
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.title('Z-Test Critical Regions')
    plt.legend()
    plt.grid(True)
    plt.show()

# User Input
sample_mean = float(input("Enter the sample mean: "))
population_mean = float(input("Enter the population mean: "))
population_std = float(input("Enter the population standard deviation: "))
sample_size = int(input("Enter the sample size: "))
alpha = float(input("Enter the significance level (e.g., 0.05 for 5%): "))

# Calculate Z-Score and critical value
z_score, critical_value = z_test(sample_mean, population_mean, population_std, sample_size, alpha)

# Plot the Z-Test
plot_z_test(z_score, critical_value, alpha)
```

```python
#normal distribution
import matplotlib.pyplot as plt
import numpy as np
#parameters
mean = 0 #represents center of distribution

std_dev = 1 #indicates spread of data
#generate data points

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000) #range of values, array with range of values ranging from (u-4q) to (u+4q)
#this range captures most of the distribution(99.7% of data in a normal distribution)

y = (1 / (std_dev *np.sqrt(2 *np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) **2) #normal distribution formula
#calculates the probability density of the normal distribution, the formula is : f(x) =(1/(qsqrt(2pi)))*e^(-(1/2)((x-u)q)^2)

#plotting the normal distribution
plt.figure(figsize =(10,6)) #set figure size for better visibility

#plot PDF curve in blue(probability density function)
plt.plot(x, y, label = 'Normal Distribution', color = 'blue')

#set title with mean and standard deviation
plt.title('Normal Distribution (Mean ={}, Std Dev ={})'.format(mean,std_dev))

#label for y axis
plt.xlabel('X-axis')

#label for x axis
plt.ylabel('Probability Density')

#verical line at mean, shown in red
plt.axvline(mean, color ='red', linestyle='--', label = 'Mean')

#vertical line at mean + q shown in green left boundary
plt.axvline(mean + std_dev, color='green',linestyle='--', label='Mean + Std Dev')

#vertical line at mean - q, also shown in green right boundary
plt.axvline(mean - std_dev, color = 'green', linestyle = '--', label='Mean - Std Dev')

#show legend to identify lines on the plot
plt.legend()

#add a grid for better readability
plt.grid()

#display plot
plt.show()
```

```python
# z testing for proportions with two tails
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test_two_proportions(x1, n1, x2, n2):
    # Calculate sample proportions
    p1 = x1 / n1
    p2 = x2 / n2
    
    # Calculate pooled proportion
    p_hat = (x1 + x2) / (n1 + n2)
    
    # Calculate the Z-score
    z_score = (p1 - p2) / np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))
    return z_score, p_hat

def plot_z_test(z_score, alpha):
    # Calculate the critical value from the normal distribution
    critical_value = norm.ppf(1 - alpha / 2)
    
    # Plot the normal distribution curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Normal Distribution')
    
    # Shade the critical regions
    plt.fill_between(x, 0, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label='Critical Regions')
    plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-Score = {z_score:.2f}')
    plt.axvline(-critical_value, color='red', linestyle='--', linewidth=2)
    plt.axvline(critical_value, color='red', linestyle='--', linewidth=2)
    
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.title('Two-Sample Z-Test for Proportions Critical Regions')
    plt.legend()
    plt.grid(True)
    plt.show()

# User Input
x1 = int(input("Enter the number of successes in sample 1 (x1): "))
n1 = int(input("Enter the sample size for sample 1 (n1): "))
x2 = int(input("Enter the number of successes in sample 2 (x2): "))
n2 = int(input("Enter the sample size for sample 2 (n2): "))
alpha = float(input("Enter the significance level (e.g., 0.05 for 5%): "))

# Calculate Z-Score and pooled proportion
z_score, p_hat = z_test_two_proportions(x1, n1, x2, n2)

# Plot the Z-Test
plot_z_test(z_score, alpha)
```
